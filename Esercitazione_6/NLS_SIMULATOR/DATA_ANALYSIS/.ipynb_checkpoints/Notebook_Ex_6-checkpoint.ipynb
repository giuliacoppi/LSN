## Ex 1

The goal of this exercise is to simulate a one-dimensional Ising model consisting of $N=50$ spins with nearest-neighbor interactions ($J=1$). In 1D, the Ising model can be solved exactly, providing a perfect benchmark for our stochastic simulations.

I implemented the Gibbs sampling algorithm in system.cpp and compared its results with the Metropolis sampling algorithm to explore the Boltzmann distribution $p(\{ s_i \}) = \exp(-\beta H)/Z$:
- **Metropolis Algorithm**: the system configuration is updated by randomly flipping a single spin. The acceptance probability is

$$
p(s \rightarrow s^\prime)=\min (1, \, e^{-\beta \Delta E})
$$

- **Gibbs Algorithm**: each move is accepted with probability 1 because the sampling is performed directly from the conditional distribution of the current spin, given the state of all other spins. The transition probability is

$$
p(s_i = \pm 1 | \{s_j\}_{j\neq i}) =\frac{1}{1 - \exp( \mp 2 \beta \sum_{j \in near} J_{ij}s_j)}
$$


The code already computed the internal energy: $U(N,T) = \langle H \rangle$. Therefore I added the calculation of $C(N,T)$, $\chi(N,T)$, $M(N,T,h)$ and computed:
1. $U(N,T) = \langle H \rangle_{h=0}$
2. $C(N,T) = k_B \beta^2 (\langle H^2 \rangle_{h=0}-\langle H \rangle_{h=0}^2)$ 
3. $\chi(N,T) = \beta \langle (\sum_{i=1,N}s_i)^2 \rangle_{h=0}$
4. $M(N,T,h=0.02)= \langle \sum_{i=1,N}s_i \rangle_{h=0.02}$

as a function of $T$ in the range $\left[0.5;2.0\right]$ (step di 0.1) for a system of $N=50$ spins with $J=1$. I started the simulation from $T=2.0$

The simulations were performed for a set of temperatures in the range $T \in [0.5, 2.0]$ with a step of $0.1$. To ensure the quality of the results each simulation started from the final configuration of the previous (higher) temperature to reduce the time needed to reach thermal equilibrium (therefore the simulations started from $T=2.0$. An initial equilibration phase was performed before starting the measurements.

I used the data blocking technique to estimate statistical uncertainties, performing a large number of Monte Carlo steps ($N_{steps} = 2 \cdot 10^5, \,\,\, N_{blocks}=20$) to ensure convergence and satisfy the Central Limit Theorem.

I computed the Internal Energy ($U$), Heat Capacity ($C$), and Magnetic Susceptibility ($\chi$) at zero external field ($h=0$), and the Magnetization ($M$) with an external field $h=0.02$.

To manage the output files, I also created two .sh files called simulation_metro.sh and simulation_gibbs.sh which automated the entire process:
- Starting the simulation from $h=0$
- Correctly initializing every parameter in input.dat and config.xyz
- Selecting the calculation of $U, C_V, M$, and $\chi$ in properties.dat
- Simulating for every $T$ (checking for critical errors), updating the configuration, moving the output files to the correct destination directory, and clearing the temporary output directory
- Updating $h$ to $0.02$, selecting only $M$ in properties.dat, and repeating the procedure
In this way, I organized every output file as a function of the simulation temperature and type.

Then, I created two additional .sh files, write_output_metro.sh and write_output_gibbs.sh, which automated the data extraction for the final results. I was interested in the last line of each output file (corresponding to the final mean and error after data blocking). These scripts iterate through every $T$ and collect the results into a single final output file. 

Thanks to these four executables, I simplified the workflow (applying techniques I learned last semester in Prof. Guerra's 'Simulations of Condensed Matter and Biosystems' course)."

The following plots show the results obtained with both Metropolis and Gibbs sampling, compared against the exact analytical curves (solid lines).
As shown in the figures, both simulation methods are in excellent agreement with the theoretical predictions across the entire temperature range. The error bars, calculated via data blocking, are small and consistent with the exact values, demonstrating that the simulator correctly samples the configuration space. While Metropolis and Gibbs use different stochastic moves, they asymptotically converge to the same thermodynamic equilibrium, confirming the robustness of the implementation.

```python
import matplotlib.pyplot as plt
import numpy as np

#TOTAL ENERGY

data_metro =  np.loadtxt("metropolis_total_energy.dat")
data_gibbs = np.loadtxt("gibbs_total_energy.dat")

x_metro = data_metro[:, 0]
y_metro = data_metro[:, 1]
err_metro = data_metro[:, 2]

x_gibbs = data_gibbs[:, 0]
y_gibbs = data_gibbs[:, 1]
err_gibbs = data_gibbs[:, 2]

points=100
T = np.linspace(0.2,3.0,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].plot(T, e, label = 'Theory', color = 'magenta')
ax[0].errorbar(x_metro, y_metro, yerr=err_metro, fmt='.', color='black', label='Metropolis Simulation')
ax[0].set_title("Ising 1D: U(N, T) , Metropolis Simulation")
ax[0].set_xlabel("T")
ax[0].set_ylabel("U(N, T)")
ax[0].legend()
ax[0].grid(True)

ax[1].plot(T, e, label = 'Theory', color = 'magenta')
ax[1].errorbar(x_gibbs, y_gibbs, yerr=err_gibbs, fmt='.', color='black', label='Gibbs Simulation')
ax[1].set_title("Ising 1D: U(N, T), Gibbs Simulation")
ax[1].set_xlabel("T")
ax[1].set_ylabel("U(N, T)")
ax[1].legend()
ax[1].grid(True)

plt.tight_layout()

```

```python
import matplotlib.pyplot as plt
import numpy as np

#SPECIFIC HEAT

data_metro =  np.loadtxt("metropolis_specific_heat.dat")
data_gibbs = np.loadtxt("gibbs_specific_heat.dat")

x_metro = data_metro[:, 0]
y_metro = data_metro[:, 1]
err_metro = data_metro[:, 2]

x_gibbs = data_gibbs[:, 0]
y_gibbs = data_gibbs[:, 1]
err_gibbs = data_gibbs[:, 2]

heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].plot(T, heat, label = 'Theory', color = 'magenta')
ax[0].errorbar(x_metro, y_metro, yerr=err_metro, fmt='.', color='black', label='Metropolis Simulation')
ax[0].set_title("Ising 1D: Heat Capacity, Metropolis Simulation")
ax[0].set_xlabel("T")
ax[0].set_ylabel("C(N, T)")
ax[0].legend()
ax[0].grid(True)

ax[1].plot(T, heat, label = 'Theory', color = 'magenta')
ax[1].errorbar(x_gibbs, y_gibbs, yerr=err_gibbs, fmt='.', color='black', label='Gibbs Simulation')
ax[1].set_title("Ising 1D: Heat Capacity, Gibbs Simulation")
ax[1].set_xlabel("T")
ax[1].set_ylabel("C(N, T)")
ax[1].legend()
ax[1].grid(True)

plt.tight_layout()


```

```python
import matplotlib.pyplot as plt
import numpy as np

#SUSCEPTIBILITY

data_metro =  np.loadtxt("metropolis_susceptibility.dat")
data_gibbs = np.loadtxt("gibbs_susceptibility.dat")

x_metro = data_metro[:, 0]
y_metro = data_metro[:, 1]
err_metro = data_metro[:, 2]

x_gibbs = data_gibbs[:, 0]
y_gibbs = data_gibbs[:, 1]
err_gibbs = data_gibbs[:, 2]

X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].plot(T, X, label = 'Theory', color = 'magenta')
ax[0].errorbar(x_metro, y_metro, yerr=err_metro, fmt='.', color='black', label='Metropolis Simulation')
ax[0].set_title("Ising 1D: $\chi$(N, T), Metropolis Simulation")
ax[0].set_xlabel("T")
ax[0].set_ylabel("$\chi$(N, T)")
ax[0].legend()
ax[0].grid(True)

ax[1].plot(T, X, label = 'Theory', color = 'magenta')
ax[1].errorbar(x_gibbs, y_gibbs, yerr=err_gibbs, fmt='.', color='black', label='Gibbs Simulation')
ax[1].set_title("Ising 1D: $\chi$(N, T), Gibbs Simulation")
ax[1].set_xlabel("T")
ax[1].set_ylabel("$\chi$(N, T")
ax[1].legend()
ax[1].grid(True)

plt.tight_layout()


```

```python
import matplotlib.pyplot as plt
import numpy as np

#MAGNETIZATION

data_metro =  np.loadtxt("metropolis_magnetization.dat")
data_gibbs = np.loadtxt("gibbs_magnetization.dat")

x_metro = data_metro[:, 0]
y_metro = data_metro[:, 1]
err_metro = data_metro[:, 2]

x_gibbs = data_gibbs[:, 0]
y_gibbs = data_gibbs[:, 1]
err_gibbs = data_gibbs[:, 2]

h=0.02 #external field
b = 1/T

l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].plot(T, M, label = 'Theory', color = 'magenta')
ax[0].errorbar(x_metro, y_metro, yerr=err_metro, fmt='.', color='black', label='Metropolis Simulation')
ax[0].set_title("Ising 1D: M(N, T) for h=0.02, Metropolis Simulation")
ax[0].set_xlabel("T")
ax[0].set_ylabel("M(N, T)")
ax[0].legend()
ax[0].grid(True)

ax[1].plot(T, M, label = 'Theory', color = 'magenta')
ax[1].errorbar(x_gibbs, y_gibbs, yerr=err_gibbs, fmt='.', color='black', label='Gibbs Simulation')
ax[1].set_title("Ising 1D: M(N, T) for h=0.02, Gibbs Simulation")
ax[1].set_xlabel("T")
ax[1].set_ylabel("M(N, T)")
ax[1].legend()
ax[1].grid(True)

plt.tight_layout()


```
