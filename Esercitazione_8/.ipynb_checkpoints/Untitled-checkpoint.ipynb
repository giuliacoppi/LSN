```python
import matplotlib.pyplot as plt
import numpy as np

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

data_test_1 = np.loadtxt("Ex/energy_test_1.dat")
data_test_2 = np.loadtxt("Ex/energy_test_2.dat")

#energy test
x_1 = data_test_1[:, 0]    #Number of blocks
y_1 = data_test_1[:, 1]    #Value
error_1 = data_test_1[:, 2]  #Standard deviation of the mean

x_2 = data_test_2[:, 0]    #Number of blocks
y_2 = data_test_2[:, 1]    #Value
error_2 = data_test_2[:, 2]  #Standard deviation of the mean

ax1.errorbar(x_1, y_1, yerr=error_1, label="test 1", color='magenta')
ax2.errorbar(x_2, y_2, yerr=error_2, label="test 2", color='royalblue')
ax1.set_xlabel("Number of blocks")
ax1.set_ylabel("$energy$")
ax1.set_title("Estimate of energy and its uncertainty")
ax1.legend()
ax1.grid()
ax2.legend()
ax2.grid()




plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

data_test_1 = np.loadtxt("Ex/acceptance_test_1.dat")
data_test_2 = np.loadtxt("Ex/acceptance_test_2.dat")

#energy test
x_1 = data_test_1[:, 0]    #Number of blocks
y_1 = data_test_1[:, 1]    #Value

x_2 = data_test_2[:, 0]    #Number of blocks
y_2 = data_test_2[:, 1]    #Value

ax1.plot(x_1, y_1, label="test 1", color='magenta')
ax2.plot(x_2, y_2, label="test 2", color='royalblue')
ax1.set_xlabel("Number of blocks")
ax1.set_ylabel("$energy$")
ax1.set_title("Estimate of energy and its uncertainty")
ax1.legend()
ax1.grid()
ax2.legend()
ax2.grid()




plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(4, 3, figsize=(15, 15), constrained_layout=True)
mu = [0.2, 0.8]
sigma = [1.5, 0.6]

for i in range(1,3):
    data_energy = np.loadtxt(f"Ex/energy_test_{i}.dat")
    data_accept = np.loadtxt(f"Ex/acceptance_test_{i}.dat")
    data_amp = np.loadtxt(f"Ex/hist_test_{i}.dat")

    blocchi = data_energy[:, 0]
    energy = data_energy[:, 2]
    error_energy = data_energy[:, 3]
    acceptance = data_accept[:, 1]
    x_amp = data_amp[:, 0]

    # Energia media
    axes[i, 0].errorbar(blocchi, energy, yerr=error_energy,  fmt='o', markersize=3, capsize=2, color='blue', label="Energia media")
    axes[i, 0].set_title(f"Simulation with $\mu$ = {mu[i]} , $\sigma$ = {sigma[i]} - Energy")
    axes[i, 0].set_xlabel("Number of blocks")
    axes[i, 0].set_ylabel(r"$\langle E \rangle$")
    axes[i, 0].legend()
    axes[i, 0].grid(True)

    # Accettanza
    axes[i, 1].plot(blocchi, acceptance, color='red')
    axes[i, 1].set_title(f"Simulation with $\mu$ = {mu[i]} , $\sigma$ = {sigma[i]} - Acceptance")
    axes[i, 1].set_xlabel("Number of blocks")
    axes[i, 1].set_ylabel("Acceptance")
    axes[i, 1].grid(True)

    # Istogramma densità di probabilità
    axes[i, 2].hist(x_amp, bins=100, density=True, color='green', alpha=0.7, edgecolor='black')
    axes[i, 2].set_title(f"Simulation with $\mu$ = {mu[i]} , $\sigma$ = {sigma[i]} - Wave function")
    axes[i, 2].set_xlabel("x")
    axes[i, 2].set_ylabel(r"$|\psi(x)|^2$")
    axes[i, 2].grid(True)

# Mostra il grafico
plt.suptitle("Analisi delle simulazioni per diverse configurazioni $(\mu, \sigma)$", fontsize=16)
plt.show()
```
